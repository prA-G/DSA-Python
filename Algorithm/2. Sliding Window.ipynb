# *Sliding Window*

Sliding window is a technique where we:
- Create a window (subarray/ substring)
- Slide it over the data
- Avoid recomputing values again n again

>Used to optimize problems from O(n^2) --> O(n)

### Real Life Analogy

Imagine you are in car window:
- You see only whats inside the window
- As the car moves, the window slides
- Old views goes out, new view comes in.

## When to use sliding Window?

- Subarrays / Substrings
- Continuous Elements
- Maximum / Minimum / Sum / Count
- Fixed size or variable size window

## Types of Sliding Window:

**Fixed Size Window** 
- Window size = k (constant)


**Variable Size Window**
- Window size changes dynamically

# Type 1 : Fixed Size Sliding Window

### Maximum Sub Subarray of size k

```python
def maxSubArray(arr, k):
    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i]   # Add next element
        window_sum -= arr[i - k]  # Remove old element
        max_sum = max(max_sum, window_sum)
    return max_sum
```

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3
maxSubArray(arr, k)
```

> Time Complexity : O(n)    & Space Complexity : O(1)

# Type 2: Variable Size Sliding Window

### Longest Subarray with sum <= k


- Expand window --> Increase Sum

- If sum > k -->  shrink window from left

```python
def longestSubArray(arr, k):
    left = 0
    curr_sum = 0
    max_len = 0

    for right in range(len(arr)):
        curr_sum += arr[right]

        while curr_sum > k:
            curr_sum -= arr[left]
            left += 1
        max_len = max(max_len, right-left+1)
    return arr[left-1:right]
```

```python
arr = [2, 1, 5, 2, 3]
k = 7
longestSubArray(arr, k)
```

# Most Important Questions

### 1. Longest Substring Without Repeating Characters

```python
def longestNonRep(s):
    seen = set()
    left = 0
    max_len = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])
        max_len = max(max_len, right-left + 1)
    return max_len
```

```python
s = "abcdabcbb"
longestNonRep(s)
```

### 2. Average of subarrays of size k

```python
def average(nums, k):
    res = []
    s = sum(nums[:k])
    res.append(s/k)

    for i in range(k, len(nums)):
        s += nums[i]
        s -= nums[i-k]
        res.append(s/k)
    return res
```

```python
nums = [1,3,2,6,-1,4,1,8,2] 
k=5
average(nums, k)
```

### 3. Find the maximum element in each window of size k.

- **Monotonic Deque**
    - Single order value with double ended queue.
    - Insert from front
    - insert from back
    - remove from front
    - remove from back

```python
from collections import deque

def maxSliding(nums, k):
    dq = deque()
    res = []

    for i in range(len(nums)):
        if dq and dq[0] == i-k:
            dq.popleft()
        
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        dq.append(i)

        if i >= k-1:
            res.append(nums[dq[0]])
    return res
```

```python
nums = [1,3,-1,-3,5,3,6,7]
k=3
maxSliding(nums, k)
```

### First negative number in every window

```python
from collections import deque

def firstNegative(nums, k):
    dq = deque()
    res = []
    
    for i in range(len(nums)):
        if nums[i] < 0:
            dq.append(i)
        if i >= k-1:
            if dq and dq[0] < i-k + 1:
                dq.popleft()
            if dq:
                res.append(nums[dq[0]])
            else:
                res.append(0)
    return res
```

```python
nums = [12,-1,-7,8,-15,30]
k = 3
firstNegative(nums, k)
```
