# *Two Pointer Technique*

It uses two variables (pointers) to traverse a data structure (usually a **String** or an **Array**) efficiently.

- Instead of using nested loops (O(n^2)), we use **Two Pointer** to reduce time to O(n).
- Mostly used on:
    - Array
    - String
    - Sorted Data.

# Types of Two Pointers

### Opposite direction
- One pointer at start
- One pointer at end
- Move inward

    - Used in:
        - Pair Sum
        - Palindrome
        - Reverse array

### Same direction
- Both start from beginning
- One moves faster
    - Used in:
        - Remove duplicates
        - Detecting cycles
        - Sliding window base.

# Why Two Pointer is SuperImportant?

- **Converts** :
    - O(n^2) >> O(n)
    - Brute Force >> Optimal
- **Asked in** :
    - FAANG
    - Leetcode
    - Codeforces
    - Real Systems
- **Common Problems** :
    - Two Sum
    - 3 sum
    - 4 sum
    - Container with most water
    - Trapping rain water
    - Remove duplicates

# Most Asked Question:

1. **When can we use two pointers?**
- When array/string is sorteed or traversal logic exists.

2. **Can we use two pointers on unsorted array?**
- Usually No, First combined with hashing or sorting

3. **Difference between sliding window and two pointer?**
- Two Pointer : Flexible movement
- SLiding Window : Fixed / Variable window size

4. **Is two pointer always O(n)?**
- Yes in most correct implementations

# Two Pointer in *Array*

### 1. Reverse an array

```python
def reverse_array(arr):
    l, r = 0, len(arr) - 1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1
        r -= 1
    return arr
```

```python
arr = [1, 2, 3, 4, 5]
reverse_array(arr)
```

> Time Complexity : O(n)  | Space COmplexity : O(1)

### 2. Check Palindrome Array

```python
def isPalindrome(arr):
    l, r = 0, len(arr) - 1
    while l < r:
        if arr[l] != arr[r]:
            return False
        l += 1
        r -= 1
    return True
```

```python
arr = [1, 2, 3, 2, 1]
isPalindrome(arr)
```

### 3. Pair with Given Sum (Sorted Array)

```python
def pairSum(arr, target):
    l, r = 0, len(arr) - 1
    while l < r:
        s = arr[l] + arr[r]
        if s == target:
            return arr[l], arr[r]
        elif s< target:
            l += 1
        else:
            r -= 1
    return -1
```

```python
arr = [1, 2, 3, 4, 5]
target = 6
pairSum(arr, target)
```

```python
# All possible pairs whose sum is equal to target:
def allPairSum(arr, target):
    l, r = 0, len(arr)-1
    res = []
    while l < r:
        s = arr[l] + arr[r]
        if s == target:
            res.append([arr[l], arr[r]])
            l += 1
            r -= 1
        elif s < target:
            l += 1
        else:
            r -= 1
    return res
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
target = 8
allPairSum(arr, target)
```

### 4. Move all Zeros to End

```python
def moveZeros(arr):
    zero = 0
    for i in range(len(arr)):
        if arr[i] != 0:
            arr[zero], arr[i] = arr[i], arr[zero]
            zero += 1
    return arr
```

```python
arr = [1,2,0,3,0]
moveZeros(arr)
```

### 5. Remove duplicates from Sorted Array

Remove duplicates from sorted array and then return Len of remaining elements of array.

```python
def removeDuplicates(arr):
    i = 0
    for j in range(1, len(arr)):
        if arr[j] != arr[i]:
            i += 1
            arr[i] = arr[j]
    return i+1
```

```python
arr = [1,2,2,3,4,4,5]
removeDuplicates(arr)
```

> Time : O(n) >> Single Pass     |      Space : O(1) >> in-place, No extra Memory

### 6. Squares of Sorted Array

```python
def sorted_squares(arr):
    for i in range(len(arr)):
        arr[i] = arr[i] * arr[i]
    arr.sort()
    return arr
```

```python
arr = [1, 2, 6, 4]
sorted_squares(arr)
```

```python
def sorted_squaresTwoPointer(arr):
    n = len(arr)
    res = [0] * n
    l, r = 0, n-1
    pos = n-1

    while l<=r:
        if abs(arr[l]) > abs(arr[r]):
            res[pos] = arr[l] *arr[l]
            l += 1
        else:
            res[pos] = arr[r]*arr[r]
            r -= 1
        pos -= 1
    return res
```

```python
arr = [-4,-1,0,3,10]
sorted_squaresTwoPointer(arr)
```

### 7. Container with most water

```python
def containerWater(arr):
    l, r = 0, len(arr)-1
    ans = 0
    while l < r:
        ans = max(ans, min(arr[l], arr[r]) * (r-l))
        if arr[l] < arr[r]:
            l += 1
        else:
            r -= 1
    return ans
        
```

```python
arr = [1,8,6,2,5,4,8,3,7]
containerWater(arr)
```

### 8. Two Sum II (Sorted Input)

```python
def twoSum(arr, target):
    l, r = 0, len(arr)-1
    while l < r:
        s = arr[l] + arr[r]
        if s == target:
            return [f"{arr[l]} + {arr[r]} = {target}"]
        elif s < target:
            l += 1
        else:
            r -= 1
```

```python
arr = [1, 2, 3, 4, 5, 6]  
target = 10
twoSum(arr, target)
```

### 9. Sort 0's & 1's

```python
def sortZeroOnes(arr):
    l, r = 0, len(arr) - 1
    while l < r:
        if arr[l] == 0:
            l += 1
        elif arr[r] == 1:
            r -= 1
        else:
            arr[l], arr[r] = arr[r], arr[l]
            l += 1
            r -= 1
    return arr
```

```python
arr = [1, 0, 1, 1, 0, 0, 1]
sortZeroOnes(arr)
```

### 10. Sort Colors (0, 1, 2) ~ Dutch National Flag

```python
def sortColors(arr):
    low, med, high = 0, 0, len(arr)-1

    while med < high:
        if arr[med] == 0:
            arr[low], arr[med] = arr[med], arr[low]
            low += 1
            med += 1
        elif arr[med] == 1:
            med += 1
        else:
            arr[high], arr[med] = arr[med], arr[high]
            high -= 1
    return arr
```

```python
arr = [0, 1, 0, 2, 1, 0, 2]
sortColors(arr)
```

### 11. Trapping Rain Water

```python
def trapWater(height):
    l, r = 0, len(height)-1
    left_max =  right_max = 0
    water = 0
    while l < r:
        if height[l] < height[r]:
            left_max = max(left_max, height[l])
            water += left_max - height[l]
            l += 1
        else:
            right_max = max(right_max, height[r])
            water += right_max - height[r]
            r -= 1
    return water
```

```python
height = [4, 2, 0, 3, 2, 5]
trapWater(height)
```

### 12. 3 Sum

Sum of three elements must be equal to zero.

```python
def threeSum(arr):
    arr.sort()
    res = []
    for i in range(len(arr)):
        if i > 0 and arr[i] == arr[i-1]:
            continue
        l, r = i+1, len(arr)-1
        while l < r:
            s = arr[i] + arr[l] + arr[r]
            if s == 0:
                res.append([arr[i], arr[l], arr[r]])
                l += 1; r -= 1
            elif s < 0:
                l += 1
            else:
                r -= 1
    return res
```

```python
arr = [1, 6, 2, 5, 8, 3, 9, -1, -2, -3]
threeSum(arr)
```

### 13. Valid Triangle

```python
def validTriangle(arr):
    arr.sort()
    count = 0
    for k in range(len(arr)-1, 1, -1):
        i, j = 0, k-1
        while i < j:
            if arr[i] + arr[j] > arr[k]:
                print(arr[i], arr[j], arr[k])
                count += j-i
                j -= 1
            else:
                i += 1
    return count
```

```python
arr = [1, 5, 2, 6, 3, 4]
validTriangle(arr)
```

### 14. Remove Elements In-PLace

```python
def removeElements(arr, val):
    slow = 0
    for fast in range(len(arr)):
        if arr[fast] != val:
            arr[slow] = arr[fast]
            slow += 1
    return arr[:slow]
```

```python
arr = [1, 2, 3, 4, 5]
val = 5
removeElements(arr, val)
```

### 15. Minimum difference Pair (Sorted)

```python
def minDiff(arr):
    arr.sort()
    l, r = 0, 1
    diff = float('inf')
    while r < len(arr):
        diff = min(diff, arr[r] - arr[l])
        l += 1
        r += 1
    return diff
```

```python
arr = [4, 9, 1, 32, 13]
minDiff(arr)
```
