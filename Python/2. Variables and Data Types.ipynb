# What is Variable?

A variable is a container used to store data.

In python, we use variables to store values so we can reuse them later.

```python
age = 20
```

- `age` : Variable Name
- `=` : Assignment Operator
- `25` : Value stored

# What is Data Types?

A data type tells python 'what kind of value it is'.

- **Common Data Types**:
    - `int` : Integer
    - `float` : Decimal Number
    - `str` : Text
    - `bool` : True/ False
    - `None` : No value

## Type Checking & Dynamic Typing:

- **Check Type Using** : `type()`

```python
x = 10
print(type(x))

y = 'Hello' 
print(type(y))
```

- **Dynamic Typing**:
    - Same variable, different data types.

```python
a = 10
a = "Ten"

print(type(a))  # Will store only last one
```

## Time and Space Complexity

- Variable Assignment : O(1) time
- Space used  = depends on data stored

1. What is a variable in Python?
