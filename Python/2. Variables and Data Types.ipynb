# What is Variable?

A variable is a container used to store data in memory.

In python, we use variables to store values so we can reuse them later.

Instead of writing values agian and again we store them in variables.

```python
age = 20
```

- `age` : Variable Name
- `=` : Assignment Operator
- `25` : Value stored

# What is Data Types?

A data type defines the type of data a variable holds.

- **Key Points** :
    - Variables store values
    - No need to declare type explicitly (Python is dynamically typed)
    - *Variable name rules*:
        - Must start with letter or `_`
        - Cannot start with number
        - No space allowed
        - Case Sensitive

- **Common Data Types**:
    - `int` : Integer
    - `float` : Decimal Number
    - `str` : Text
    - `bool` : True/ False
    - `None` : No value

## Type Checking & Dynamic Typing:

- **Check Type Using** : `type()`

```python
x = 10
print(type(x))

y = 'Hello' 
print(type(y))
```

- **Dynamic Typing**:
    - No need to declare data type explicitly.

    - Same variable, different data types.

```python
a = 10
a = "Ten"

print(type(a))  # Will store only last one
```

## NUMBER

#### OPERATIONS :
 
- **ADDITION** : `+`
- **SUBTRACTION** : `-`
- **MULTIPLICATION** : `*`
- **DIVISION** : `/`
- **FLOAT DIVISION** : `//`
- **REMAINDER / MODULUS** : `%`
- **POWER** : `**`

```python
a = 10
b = 3
print(f"Addition of {a} & {b} is : ", a+b)
print(f"Subtraction of {a} & {b} is : ", a-b)
print(f"Multiplication of {a} & {b} is : ", a*b)
print(f"Division of {a} & {b} is : ", a/b)
print(f"Float Division of {a} & {b} is : ", a//b)
print(f"Remainder of {a} & {b} is : ", a%b)
print(f"Power of {a} to {b} is : ", a**b)
```

- Python is following BODMAS, but for standard code and best precision we use paranthesis for clean and understandable code.


```python
x, y, z = 2, 3, 4
print(x + y * z)    # Wrong approach
print((x + y) * z)  # Correct approach
```

- Always value higher precision, but explicitely we should care about data type in which we want value.

```python
print(40 + 20.23)   # Not good approach
print(40 + int(20.23))  # Good approach
print(float(40) + 20.23)    # Good approach
```

- If we call any assign value it will store in Tuple because assigned value is not mutable.

```python
x, y, z # output in tuple
```

#### Comparison Value

When we compare any value we will get output in `Boolean`
- **True** ~ 1
- **False** ~ 0

- **Smaller than** : `<`
- **Greater than** : `>`
- **Smaller than or equal** : `<=`
- **Greater than or equal** : `>=`
- **Equal to** : `==`
- **Not equal to** : `!=`

```python
a = 5
b = 3
print(f"Is {a} smaller than to {b}? :", a<b)
print(f"Is {a} greater than to {b}? :", a>b)
print(f"Is {a} smaller than or equal to {b}? :", a<=b)
print(f"Is {a} greater than or equal to {b}? :", a>=b)
print(f"Is {a} equal to {b}? :", a==b)
print(f"Is {a} not equal to {b}? :", a!=b)
```

#### Types of number:

- **Binary** : base 2 >> (0 & 1)
- **Octal** : base 8 >> (0 to 7)
- **Decimal** : base 10 >> (0 to 9)
- **HexaDecimal** : base 16 >> (0 to 9 & A to F)

```python
# from Octal to decimal
print(0o20)
print(0o26)
```

```python
# from decimal to octal
print(oct(16))
print(oct(22))
```

```python
# from hexadecimal to decimal
print(0xFF)
print(0x24)
```

```python
# from decimal to hexadecimal
print(hex(255))
print(hex(36))
```

```python
# from binary to decimal
print(0b1001)
print(0b11001)
```

```python
# from decimal to binary
print(bin(9))
print(bin(25))
```

- We can use int() for getting any base value by mention it.
- `int("num", base)`

```python
print(int('11001', 2)) # base 2 for binary
print(int('20', 8)) # base 8 for octal
print(int('64', 16)) # base 16 for hexadecimal
```

## Advanced Data Types

- 1. **LIST** : An unordered collection of data with element seperated by comma and enclosed with square bracket. Lists are **mutable** which means can be changed after creations.

- 2. **TUPLE** : An ordered collection of data with elements separated be a cpmma and enclosed with parenthesis. Tuples are **immutbale** and can not be changed after creations.

- 3. **Dictionary** : A dictionary is an unordered colection of data containing a **key-value** pair. These are enclosed within curly braces.

```python
# List
a = [1, 3.5, "abc"]
print(a)
```

```python
a[1] = 45.55
print(f"Modified List {a}")
print(type(a))
```

```python
#Tuple
b = (2, 3.6, "def")
print(type(b))
```

```python
try:
    b = (10, 20, 30)
    print(f"Original Tuple {b}")

    b[1] = 40
    print(f"Modified Tuple {b}")
except Exception as e:
    print("Error : ", e)
```

```python
#Dict
Dictss = {'name':'ABC', 'class':2, 'present':True}
print(Dictss)
print(type(Dictss))
```
