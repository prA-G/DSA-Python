# Conditional Statement

A conditional statement allows a program to make decisions.

- It checks a condition
- If the condition is True, some code runs
- If False, other code runs

- **In Python we use**
    - `if`
    - `elif`
    - `else`
    

# If Statement

```python
if condition:
    code
```

````python
age = 20
if age >= 18:
    print("You can vote")
````

> Indentation Space is mandatory.

# If- Else statement

When there are two possibilities.

````python
age = 16
if age>= 18:
    print("eligible to vote")
else:
    print("Not eligible")
````

# If-Elif-Else statement

Used for multiple conditions

````python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
````

# Nested-if 

If- else inside if-else

````python
age = 15
has_id = True

if age >= 18:
    if has_id:
        print("Entry allowed")
    else:
        print("ID required")
else:
    print("Underage")
````

## Using logical operators in conditions

````python
age = 22
citizen = True

if age >= 18 and citizen:
    print("Can vote")
else:
    print("Cannot vote")
````

## Common Interview Questions

### 1. Difference between `if` and `elif`?
- `if` starts condition, `elif` checks another condition if previous was false.

### 2. Can we have multiple else?
- No, only one else per if block.

### 3. Is `elif` mandatory?
- No, you can use only `if` or `if-else`.

### 4. What happens if no condition is True?
- The else block executes (if present)

### 5. Why is indentation important?
- Python uses indentation to define code blocks


# Practice problem

1. WAP to check whether a number is Positive/ Negative or Zero.

````python
num = int(input("Enter a number: "))
print("Given Number is : ", num)
if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
````

2. WAP to check a number is even or odd?

````python
num = int(input("Enter a number: "))
print("Enter the number : ", num)
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
````

3. Find the largest among three numbers.

````python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
print("1st number : ", a)
print("2nd number : ", b)
print("3rd number : ", c)
if a >= b and a >= c:
    print("Largest is:", a)
elif b >= a and b >= c:
    print("Largest is:", b)
else:
    print("Largest is:", c)
````

4. Check if a year is leap year?

````python
year = int(input("Enter a year: "))
print("Enter th year : ", year)
if year % 400 == 0:
    print("Leap year")
elif year % 100 == 0:
    print("Not a leap year")
elif year % 4 == 0:
    print("Leap year")
else:
    print("Not a leap year")
````

5. Make a calculator using if-elif.

````python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")
print("Enter first number : ", a)
print("Enter second number : ", b)
print("Enter operator from +, -, *, / :", op)
if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
else:
    print("Invalid operator")
````

### Important question

- Check whether a number is :
    - divisible by both 3 and 5
    - only by 3
    - only by 5
    - neither.

````python

def checkDiv(num):
    if num%15==0:
        print(f"{num} is divisible by both 3 and 5")
    elif num%3==0:
        print(f"{num} is divisible by 3 only")
    elif num%5==0:
        print(f"{num} is divisible by 5 only")
    else:
        print(f"{num} is not divisible")

````

````python
num = 45
checkDiv(num)
````

````python
num = 27
checkDiv(num)
````
