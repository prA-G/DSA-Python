# Loop


A loop is used to repeat a block of code multiple times.

- **Why loop exists?**
    - Avoid writing same code again and again
    - Work with lists, strings, ranges
    - Perform repetitive tasks efficiently

# Types of Loop

## For Loop

A for loop is used when:
- Number of repetitions is known
- Iterating over a sequence (list, string, range)

### Syntax

```python
for variable in sequence:
    code

- Example : Print 1 to 5

````python
for i in range(1, 6):
    print(i)
````

### `range()` Function

range(start, stop, step)

## While loop

A while loop :
- Runs as long as condition is True
- Number of iteration is unknown

### Basic Syntax

```python
while condition:
    code

- Example : Print 1 to 5

````python
i = 1
while i <= 5:
    print(i)
    i += 1
````

## Difference between For and while loop

| Feature           | for loop      | while loop       |
| ----------------- | ------------- | ---------------- |
| Iterations known? | Yes           | No               |
| Used with         | Sequence      | Condition        |
| Safer             | Yes           | Risk of infinite |
| Common use        | Lists, ranges | User input       |


### Practice problem

````python
# Print all number from 10 to 1 using while

i = 10
while i >= 1:
    print(i)
    i -= 1
````

````python
# Print all odd numbers from 1 to 20 usinf for

for i in range(1, 20, 2):
    print(i)
````

````python
# Sum of digits

def sumDigit(num):
    sum_digit = 0

    while num > 0:
        digit = num % 10
        sum_digit+= digit
        num//=10
    return sum_digit
````

````python
num = 1234
sumDigit(num)
````
