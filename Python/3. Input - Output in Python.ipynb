# Output in Python

Output means showing data to the user.

Python uses `print()` to show output.

```python
print("Name : ABC")
```

# Input in Python

Input means taking data from the user.

Python uses `input()` to take data from the users.

```python
name = input("Enter your name : ")
print("Hello", name)
```

### Key Points:

- `input()` always returns string type
- Use type conversion if needed:
    - `int()`, `float()`
- `print()` can display multiple values
- Formatting output improves readability.

```python
# Example 
name = input("Enter your Name : ")
print("Hello", name)

age = int(input("Enter your age : "))
print("Your age is ", age)
```

# Edge Cases

- **`input()` always gives string**

```python
x = input()
print(type(x))
```

- **Must converted when needed**

```python
num = int(input())
print(type(num))
```

- **Invalid conversion**

```python
int("abc")
```

- **Multiple inputs**

```python
a, b = input().split()
print(a, b)
```

- *Best Practice*

```python
nums = list(map(int, input().split()))
print(nums)
```

- `split()` returns List of strings

# Type Casting

Type Casting means converting one data into another data type.

- Python supports a wide variety of functions or methods like : `int()`, `str()`, `ord()`, `hex()`, `typle()`, `set()`, `list()`, etc, for typecasting in Python.

- **Key Points**
    - Python supports dynamic typing
    - Two types of casting:
        - Implicit Casting (Automatic)
        - Explicit Casting (Manual)
    - Common Functions:
        - `int()` > Convert to integer
        - `str()` > Convert to string
        - `float()` > Convert to float
        - `bool()` > COnvert to boolean

### Two Types of Type Casting

1. **Implicit Type Casting**

Python automatically converts data types when needed.
    

```python
# Example 
x = 10      # int
y = 5.5     # float

result = x+y
print(result, ", And its type is : ", type(result))
```

It converts smaller type to larger type.

> int -->> float -->> complex

2. **Explicit Type Casting**

Manual conversion using functions like `int()`, `float()`, etc

```python
# Example of explicit conversion

string = "14"
number = int(string)

print(number, ", And its type is ", type(number))
```

### Common Type Casting example

```python
# String -->> Integer

s = "100"
i = int(s)
print(f"{i} and its type is {type(i)}")
```

```python
# Int -->> Float

i = 10
f = float(i)
print(f"{i} and its type is {type(i)}\n{f} and its type is {type(f)}")
```

```python
# Float -->> Int
f = 10.2
i = int(f)  # remove decimal part
print(f"{f} and its type is {type(f)}\n{i} and its type is {type(i)}")
```

```python
# Int -->> String

i = 100
s = str(i)
print(f"{i} and its type is {type(i)}\n{s} and its type is {type(s)}")
```

```python
# Boolean Conversion
print(bool(1))
print(bool(0))
print(bool(""))
print(bool("Hiii"))

# for 0 and Empty value it show False only
```

### Common Interview Questions

1. What is type Casting?
    - Converting one data type into another
2. Difference between implicit and explicit casting?
    - Implicit = Automatic & Explicit = Manual
3. What happens in int(5.9)?
    - Decimal part is removed not rounded.
4. Can we convert "abc" into `int()`?
    - No, it throws ValueError.
5. Why does `int("1, ")` fail?
    - Because it is not a valid integer string
6. Default behavior of `split()`?
    - Split by space by default
7. How to handle comma-separated input?
    - `split(",")` or clean string first
