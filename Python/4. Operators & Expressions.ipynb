# Operator

An operator is a symbol that performs an operation on values (called operand).


# Expression

An expression is a combination of values + operators that produces a result.

- 2 + 3 : `+` >> Operator and `2` , `3` >> Operands
- `10*2+5` >> Expression

# Types of operators:

## Arithmetic Operator

- Used for mathematical calculations:

| Operator | Meaning             | Example  |
| -------- | ------------------- | -------- |
| `+`      | Addition            | `5 + 2`  |
| `-`      | Subtraction         | `5 - 2`  |
| `*`      | Multiplication      | `5 * 2`  |
| `/`      | Division            | `5 / 2`  |
| `//`     | Floor Division      | `5 // 2` |
| `%`      | Modulus (remainder) | `5 % 2`  |
| `**`     | Power               | `5 ** 2` |


### Example

```python
a = 10
b = 3

print("Addition : ", a + b)
print("Subtraction : ", a - b)
print("Multiplication : ", a * b)
print("Division : ", a / b)
print("Float division : ", a // b)
print("Modulus Remainder : ", a % b)
print("Power : ", a ** b)
```

## Comparison Operator

- Used to compare values >> Result is True or False

| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | Equal            |
| `!=`     | Not equal        |
| `>`      | Greater than     |
| `<`      | Less than        |
| `>=`     | Greater or equal |
| `<=`     | Less or equal    |


```python
x = 5
y = 10

print(x == y)
print(x < y)
print(x != y)
```

## Logical Operators

- Used to combine conditions

| Operator | Meaning           |
| -------- | ----------------- |
| `and`    | Both true         |
| `or`     | At least one true |
| `not`    | Reverse result    |


```python
age = 20

print(age > 18 and age < 30)
print(age < 18 or age > 60)
print(not age < 18)
```

## Common Interview Questions:

1. Difference between `/` and `//` ?
- `/` gives float result, `//` gives integer (floor value)

2. What is operator precedence?
- It decides which operator executes first in an expression.

3. What does `%` operator do?
- It returns the remainder after division.

4. Difference between `==` and `=` .
- `=` assign value and `==` compares value
