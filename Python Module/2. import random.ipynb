This module helps to get any random value between 0 to 1 by default and can customize also fr any number , range , or list.

```python
import random
```

```python
print(random.random())
print(random.random())
print(random.random())
print(random.random())
print(random.random())
print(random.random())
```

### random.randint()
- We can assign range between which we want random values.

```python
print(random.randint(1, 100))
print(random.randint(1, 100))
print(random.randint(1, 100))
print(random.randint(1, 100))
print(random.randint(1, 100))
print(random.randint(1, 100))
```

### random.choice()

- We can assign any array or list and we can randomly get value from that assigned list /  array for getting random choices.

```python
l1 = ['Python', 'C++', 'Java', 'React', 'Odoo']
print(random.choice(l1))
print(random.choice(l1))
print(random.choice(l1))
print(random.choice(l1))
print(random.choice(l1))
```

### random.shuffle()
- We can use this to shuffle any assigned list.

```python
l2 = ['heart', 'spade', 'diamond', 'club']
random.shuffle(l2)
print("Shuffle 1 : ", l2)
random.shuffle(l2)
print("Shuffle 2 : ",l2)
random.shuffle(l2)
print("Shuffle 3 : ",l2)
random.shuffle(l2)
print("Shuffle 4 : ",l2)
```
