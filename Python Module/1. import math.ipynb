This module helps to perform mathematical calculations in many ways.

```python
import math
```

#### math.floor

This will give bottom value to any decimal number (according to number line).

```python
print(math.floor(3.6))
print(math.floor(-3.6))
print(math.floor(8.2))
print(math.floor(-8.2))
```

#### math.trunc()

This will give value which is towards zero of any decimal number (according to number line)

```python
print(math.trunc(3.6))
print(math.trunc(-3.6))
print(math.trunc(8.2))
print(math.trunc(-8.2))
```
