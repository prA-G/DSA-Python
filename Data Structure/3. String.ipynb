# String

A string is a sequence of characters stored contiguously.

```python
s = "Hello"
s
```

```python
print(type(s))
```

## Important Properties :

- Strings are indexed (like arrays)
- Strings are immutable in Python  >> Cannot change character directly once created.

## Real Life Analogy:

Think of a string like a printed word on paper.
- Can read characters easily.
- But to change a letter, you must re-write the whole word.  >> Immutability.

# Basic Operations

```python
s = "Delulu World"
```

```python
# Access characters of string
print(s[0])  # 1st character
print(s[-2])  # 2nd last character
```

```python
# Length of string
print(len(s))  # Number of characters including spaces
```

```python
# Slicing
print(s[0:5])  # characters from index 0 to 4
print(s[:7])  # characters from start to index 6
print(s[6:])  # characters from index 6 to end
```

```python
# Concatenation
s1 = "I Love "
print(s1 + s)
```

```python
# Membership 
print('orld' in s)
print('Mahi' in s)
```

```python
# Reverse string
print(s[::-1])
```

# Important Questions

### 1. Reverse the given string

```python
def reverseString(s):
    return s[::-1]
```

```python
s = "AbcDefGhi"
reverseString(s)
```

### 2. Check Palindrome 

```python
def isPalindrome(s):
    left, right = 0, len(s)-1

    while left < right:
        return s[left] == s[right]
        left += 1
        right -= 1
    return True
```

```python
s = "madam"
isPalindrome(s)
```

```python
s = "pragati"
isPalindrome(s)
```

### 3. Valid Anagram

Same length and same frequency (Order does not matter)

```python
def isAnagram(a, b):
    return len(a) == len(b)

    freq = {}
    for ch in a:
        freq[ch] = freq.get(ch, 0) + 1

    for ch in b:
        if ch not in freq or freq[ch] ==0:
            return False
        freq[ch] -= 1
    return True
```

```python
a = "anagram"
b = 'mgrnaaa'
isAnagram(a, b)
```

### 4. First non-repeating character

```python
def firstUnique(s):
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    for i in range(len(s)):
        if freq[s[i]] == 1:
            return s[i]
    return -1
```

```python
s = "aabcddef"
firstUnique(s)
```

```python
# 2nd non-repeating character

def secNonRep(s):
    freq = {}
    res = []
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    for i in range(len(s)):
        if freq[s[i]] == 1:
            res.append(s[i])
    if len(res) >= 2:
        return res[1]
    return -1
```

```python
a = "abbccddeff"
secNonRep(a)
```

```python
a = "leettccooddee"
secNonRep(a)
```

### 5. Longest Substring without repeating characters

**Sliding Window**

```python
def lenOfLongestString(s):
    seen = set()
    left = 0
    max_len = 0
    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])
        max_len = max(max_len, right - left + 1)

    return max_len
```

```python
s = "leetcode"
lenOfLongestString(s)
```

```python
# Print longest substring without repeating character (not length)

def nonRepSubstring(s):
    seen = set()
    left = 0
    
    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])
    return s[left : right+1]
```

```python
s = "leetccoders"
nonRepSubstring(s)
```
