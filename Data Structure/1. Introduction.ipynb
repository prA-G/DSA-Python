# DSA (Data Sturcture & Algorithms)

- **Data Structure** : How data is stored & organized?
- **ALgorithm** : Steps to solve a problem

Together they helps us to write fast, efficient, scalable code.

- Example :
    - Data : Number of students
    - Structure : Array/ List
    - Algorithm : Find topper ( Max)

## Time & Space Complexity 

- **Time Complexity** : O(n) >> We check every element once
- **Space Complexity** : O(1) >> No extra memory used

# Important Questions:

1. Why should we learn DSA?
To write efficient code and crack interviews.

2. What is the difference between data structure and algorithm?
Structure stores data, algorithm processes data.

3. Is DSA language dependent?
No, concepts are same in all language.

4. What comes first: DSA or Coding?
Basics of coding -> DSA -> Advanced problems.
